# Let's compute part (b): Einstein radius for given distances and a solar mass lens

# Constants
G = 6.67430e-11  # m^3 kg^-1 s^-2
c = 299792458    # m/s
M_sun = 1.989e30  # kg
parsec = 3.0857e16  # meters
kpc = 1e3 * parsec

# Distances
D_S = 50 * kpc  # Distance to source
D_L = 10 * kpc  # Distance to lens
D_SL = D_S - D_L

# Einstein radius in radians
theta_E_rad = np.sqrt((4 * G * M_sun / c**2) * (D_SL) / (D_L * D_S))

# Convert to arcseconds
radian_to_arcsec = 206265
theta_E_arcsec = theta_E_rad * radian_to_arcsec
theta_E_arcsec
